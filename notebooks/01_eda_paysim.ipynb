# PaySim EDA

Cleaned-up version of the original `test.ipynb`. Uses the project's `src` modules
so the same loaders/feature engineering used in training are exercised here too.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from src.data.loader import load_paysim, summarize
from src.data.feature_engineering import paysim_features

sns.set_theme(style='whitegrid')
```

```python
df = load_paysim()
summarize(df, 'PaySim')
```

```python
df['type'].value_counts().plot(kind='bar', title='Transaction types', figsize=(8, 4))
plt.show()
```

```python
# Fraud rate by transaction type
df.groupby('type')['isFraud'].mean().sort_values(ascending=False)
```

```python
engineered = paysim_features(df.sample(200_000, random_state=42))
engineered[['amount_to_balance_ratio', 'balance_delta_orig', 'amount_equals_orig_balance']].describe()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(
    engineered.select_dtypes(include='number').corr(),
    annot=True, fmt='.2f', cmap='coolwarm', center=0,
)
plt.title('Correlation matrix (engineered features)')
plt.show()
```
