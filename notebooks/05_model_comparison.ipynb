# Model comparison & hypothesis test

Compares RandomForest, IsolationForest, and the rule-based baseline using the
metrics required by the research framework. Tests H1 vs HA via McNemar's test.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import json
import pandas as pd

with open(ROOT / 'models_local' / 'metadata.json') as fh:
    metadata = json.load(fh)
```

```python
rows = []
for label, m in [
    ('RandomForest (PaySim)', metadata['paysim']['rf']),
    ('Rule-based (PaySim)', metadata['paysim']['rule_based']),
    ('IsolationForest (Custom)', metadata['custom']['isolation_forest']),
    ('Rule-based (Custom)', metadata['custom']['rule_based']),
]:
    rows.append({
        'model': label,
        'accuracy': m['accuracy'], 'precision': m['precision'],
        'recall': m['recall'], 'f1': m['f1'],
    })
pd.DataFrame(rows)
```
